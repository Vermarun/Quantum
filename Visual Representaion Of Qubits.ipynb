# Overview of Quantum Gates

Link : https://quantum-computing.ibm.com/support/guides/gate-overview?section=5d00d964853ef8003c6d6820#

```python
from qiskit.tools.visualization import plot_bloch_multivector
from qiskit.tools.visualization import plot_histogram
```

## X gate

The Pauli X gate has the property of flipping the |0⟩ state to |1⟩, and vice versa. It is equivalent to Rx for the angle π.

```python
circuit = QuantumCircuit(1,1)

# X Gate
circuit.x(0)
```

## State vector representaion

```python
# We are using statevector simulator. State vector is the vector that 
# describe the state of the qbits thats why its called statevector simulator

simulator = Aer.get_backend('statevector_simulator')
```

```python
result = execute(circuit, backend=simulator).result()

statevector = result.get_statevector()
```

```python
print(statevector)

plot_bloch_multivector(statevector)
```

## By executing measurements

```python
circuit.measure(0,0)

backend = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=backend, shots=1000).result()

counts = result.get_counts()
```

```python
plot_histogram(counts)
```

```python
circuit.draw(output='mpl')
```

## H gate

The H or Hadamard gate rotates the states |0⟩ and |1⟩ to |+⟩ and |−⟩, respectively. It is useful for making superpositions. As a Clifford gate, it is useful for moving information between the x and z bases.

```python
circuit = QuantumCircuit(1,1)

# H Gate
circuit.h(0)
simulator = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend=simulator).result()
statevector = result.get_statevector()
print(statevector)
plot_bloch_multivector(statevector)
```

```python
circuit.measure(0,0)

backend = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=backend, shots=1000).result()

counts = result.get_counts()
plot_histogram(counts)
```

```python
circuit.draw(output='mpl')
```
